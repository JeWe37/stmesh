```python
import itk
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = itk.imread('data/Pat01_lvc.mhd')
```

```python
arr = itk.array_from_image(image)
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(arr[7, :, :, :])
ax.set_xlim(0, 100)
```

```python
padded = np.pad(arr, 10)
```

```python
padded_image = itk.image_from_array(padded)
padded_image.SetSpacing(image.GetSpacing())
```

```python
itk.imwrite(padded_image, 'data/padded.mha')
```
