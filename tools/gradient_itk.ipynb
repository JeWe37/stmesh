```python
import itk
import numpy as np
import matplotlib.pyplot as plt
```

```python
np_img = np.indices((20, 20, 20, 20))[2].astype(np.float32)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(np_img[12, :, 12, :], cmap='gray')
ax2.imshow(np_img[12, :, :, 12], cmap='gray')
```

```python
image = itk.image_from_array(np_img)
image.SetSpacing([1.0, 1.0, 1.0, 2.0])
```

```python
itk.imwrite(image, 'gradient.mha')
```
